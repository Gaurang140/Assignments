**Q2. What are the rules to create variables in Python?**


- **Ans:** In Python, variables are named identifiers used to store data values. The rules for creating variables are as follows:

1. Variable names must start with a letter (a-z, A-Z) or an underscore (_).
2. The remaining characters in the variable name can be letters, numbers (0-9), or underscores.
3. Variable names are case-sensitive, which means `myVar` and `myvar` are considered different variables.
4. Python keywords (reserved words) cannot be used as variable names.
5. Variable names should be descriptive and relevant to their purpose for code readability.
6. Avoid using special characters or spaces in variable names.
7. Variable names cannot start with a number but can include numbers after the first character.

**Example of valid variable names:**
- `age`
- `_name`
- `user123`
- `myVar`

**Example of invalid variable names:**
- `1user` (starts with a number)
- `my-var` (contains a hyphen)
- `for` (Python keyword)
- `my variable` (contains a space)

**Q7. Give an example of the following conditions:**
- (i) Homogeneous list
- (ii) Heterogeneous set
- (iii) Homogeneous tuple

- **Ans:**

    **(i) Homogeneous list:**
    A homogeneous list contains elements of the same data type. Here's an example of a homogeneous list of integers:

    ```python
    homogeneous_int_list = [1, 2, 3, 4, 5]
    ```

    **(ii) Heterogeneous set:**
    A heterogeneous set contains elements of different data types. Here's an example of a heterogeneous set containing strings, integers, and floats:

    ```python
    heterogeneous_set = {'apple', 123, 3.14, 'banana'}
    ```

    **(iii) Homogeneous tuple:**
    A homogeneous tuple contains elements of the same data type. Here's an example of a homogeneous tuple of strings:

    ```python
    homogeneous_str_tuple = ('apple', 'banana', 'orange', 'grape')
    ```

**Q8. Explain the mutable and immutable data types with proper explanation & examples.**


- **Ans:**



    Mutable data types can have their values changed after creation. For example, lists can be modified by adding, removing, or changing elements:

    ```python
    my_list = [1, 2, 3]
    my_list[0] = 10  # Modifying element at index 0
    ```

    Immutable data types, like strings or integers, can't be changed once created. Instead, you create new objects:

    ```python
    my_string = "Hello"
    new_string = my_string + " World"  # Creating a new string
    ```

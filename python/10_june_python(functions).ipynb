**Q1. In Python, what is the difference between a built-in function and a user-defined function? Provide an example of each.**
- **Ans:** A built-in function is provided by Python itself and can be used without creating it. Example: `len()` to get the length of a list. A user-defined function is created by the programmer. Example: 

Built-in function:
```python
length = len([1, 2, 3])
```

User-defined function:
```python
def add(a, b):
    return a + b
result = add(5, 3)
```

**Q2. How can you pass arguments to a function in Python? Explain the difference between positional arguments and keyword arguments.**
- **Ans:** Arguments are values passed to a function. Positional arguments are passed based on the position. Keyword arguments are passed with their names. 

Example of positional argument:
```python
def greet(name, message):
    print(f"Hello, {name}! {message}")
greet("Alice", "How are you?")
```

Example of keyword argument:
```python
greet(message="How are you?", name="Alice")
```

**Q3. What is the purpose of the return statement in a function? Can a function have multiple return statements? Explain with an example.**
- **Ans:** The `return` statement sends a value back from a function. Yes, a function can have multiple `return` statements. 

Example:
```python
def absolute_value(x):
    if x < 0:
        return -x
    else:
        return x
result = absolute_value(-5)
```

**Q4. What are lambda functions in Python? How are they different from regular functions? Provide an example where a lambda function can be useful.**
- **Ans:** Lambda functions are small, anonymous functions defined with the `lambda` keyword. They are used for short operations. Example:

Lambda function:
```python
double = lambda x: x * 2
result = double(5)
```

Regular function:
```python
def double(x):
    return x * 2
result = double(5)
```

**Q5. How does the concept of "scope" apply to functions in Python? Explain the difference between local scope and global scope.**
- **Ans:** Scope defines where a variable can be accessed. Local scope is within a function and can't be accessed outside. Global scope is outside any function and can be accessed everywhere.

- **Ans:** Python passes arguments by value, but for mutable objects like lists, it feels like pass by reference because changes inside the function affect the original object.

**Q8. Create a function that can intake an integer or decimal value and perform logarithmic, exponential, power functions with base 2, and square root operations.**
```python
import math

def math_operations(x):
    log_result = math.log(x)
    exp_result = math.exp(x)
    power_result = 2 ** x
    sqrt_result = math.sqrt(x)
    return log_result, exp_result, power_result, sqrt_result
```

**Q9. Create a function that takes a full name as an argument and returns first name and last name.**
```python
def split_name(full_name):
    first_name, last_name = full_name.split()
    return first_name, last_name
```

14. **What is the purpose of the `try` clause? What is the purpose of the `except` clause?**

- Ans : Ans: The try clause is used to enclose the code that might raise an exception. The purpose of the try clause is to catch exceptions and allow the program to continue running even if an error occurs. The except clause is used to specify the code that should be executed if a specific exception occurs within the try block.

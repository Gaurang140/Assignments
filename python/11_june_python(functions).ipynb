**Q1. What is a lambda function in Python, and how does it differ from a regular function?**
- **Ans:** A lambda function is an anonymous, one-liner function defined using the `lambda` keyword. It's used for short operations. Unlike regular functions, lambda functions don't require a `return` statement, and they can't have complex logic or multiple expressions.
---
---
**Q2. Can a lambda function in Python have multiple arguments? If yes, how can you define and use them?**
- **Ans:** Yes, a lambda function can have multiple arguments. Syntax: `lambda arg1, arg2: expression`. Example: `sum = lambda x, y: x + y`.
---
---
**Q3. How are lambda functions typically used in Python? Provide an example use case.**
- **Ans:** Lambda functions are often used for quick, simple operations, like sorting or filtering. Example: `sorted_list = sorted(my_list, key=lambda x: len(x))`.
---
---
**Q4. What are the advantages and limitations of lambda functions compared to regular functions in Python?**
- **Ans:** Advantages: Concise, no need for a separate function definition. Limitations: Limited to single expressions, not suitable for complex logic or large functions.
---
---
**Q5. Are lambda functions in Python able to access variables defined outside of their own scope? Explain with an example.**
- **Ans:** Yes, lambda functions can access variables defined outside their scope. Example: `x = 10; func = lambda y: x + y`.

**Q11. Write a recursive function to calculate the factorial of a given positive integer.**
```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

**Q12. Implement a recursive function to compute the nth Fibonacci number.**
```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

**Q13. Create a recursive function to find the sum of all the elements in a given list.**
```python
def sum_list(lst):
    if not lst:
        return 0
    return lst[0] + sum_list(lst[1:])
```

**Q14. Write a recursive function to determine whether a given string is a palindrome.**
```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    return s[0] == s[-1] and is_palindrome(s[1:-1])
```

**Q15. Implement a recursive function to find the greatest common divisor (GCD) of two positive integers.**
```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)
```

**Q1. Create two `int` type variables, apply addition, subtraction, division, and multiplication and store the results in variables. Then print the data in the following format by calling the variables:**
```markdown
First variable is __ & second variable is __.
Addition: __ + __ = __
Subtraction: __ - __ = __
Multiplication: __ * __ = __
Division: __ / __ = __
```

**Ans:**

```python
# Create two integer variables
num1 = 10
num2 = 5

# Perform arithmetic operations and store results
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

# Print the results in the required format
print(f"First variable is {num1} & second variable is {num2}.")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
```

**Q2. What is the difference between the following operators:**
(i) ‘/’ & ‘//’
(ii) ‘**’ & ‘^’

**Ans:**

(i) The `'/'` operator is used for division in Python and returns a floating-point result. The `'//'` operator, known as floor division, also performs division but truncates the decimal part, giving an integer result.

(ii) The `'**'` operator is used for exponentiation in Python, where `x ** y` means x raised to the power of y. The `'^'` operator, however, is not an exponentiation operator in Python. Instead, it's the bitwise XOR operator, used for performing bitwise exclusive OR on the binary representations of numbers.

**Q3. List the logical operators.**

**Ans:** The logical operators in Python are:
- `and`: Returns `True` if both operands are true.
- `or`: Returns `True` if at least one operand is true.
- `not`: Returns the opposite of the operand's truth value.

**Q4. Explain right shift operator and left shift operator with examples.**

**Ans:** The right shift operator (`>>`) shifts the bits of a number to the right by a specified number of positions. The left shift operator (`<<`) shifts the bits to the left. Here are examples:

```python
# Right shift
num = 8  # Binary: 1000
result = num >> 1  # Binary after right shift by 1: 0100 (Decimal 4)

# Left shift
num = 4  # Binary: 0100
result = num << 2  # Binary after left shift by 2: 1000 (Decimal 8)
```

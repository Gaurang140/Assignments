
---

4. **What is the value of `spam[-1]`?**
   - **Ans:** Last element of the list which is 10 here.

---

5. **What is the value of `spam[:2]`?**
   - **Ans:** Value of `spam[:2]` is [2,4] it will extract first two elements from the list and skip 3rd one.

---

6. **What is the value of `bacon.index('cat')`? (Let's pretend `bacon` has the list `[3.14, 'cat', 11, 'cat', True]`.)**
   - **Ans:**  Value of `bacon.index('cat') will be 1

---

7. **How does `bacon.append(99)` change the look of the list value in `bacon`?**
   - **Ans:** it appends the  99 value as last place of the list. bacon.append(99) would change bacon to [3.14, 'cat', 11, 'cat', True, 99].

---

8. **How does `bacon.remove('cat')` change the look of the list in `bacon`?**
   - **Ans:**   bacon.remove('cat') would change bacon to [3.14, 11, 'cat', True, 99]

---

9. **What are the list concatenation and list replication operators?**
   - **Ans:**  List concetenation operator is ```+``` and replication operatopn is ```*```
   For Example :

   ```python
   
                  list1 = [1, 2, 3]
                  list2 = [4, 5, 6]
                  concatenated_list = list1 + list2
                  # Result: [1, 2, 3, 4, 5, 6]


                  original_list = [0, 1]
                  replicated_list = original_list * 3
                  # Result: [0, 1, 0, 1, 0, 1]


   
   ```

---

10. **What is the difference between the list methods `append()` and `insert()`?**
    - **Ans:** Both methods are use to add element to a list. append method use to add element in last position of list whereas , insert method used to add element at a specific index within list.

    For example:

    ```python
                     my_list = [1, 2, 3]
                     my_list.append(4)
                     # Result: [1, 2, 3, 4]

                     my_list_2 = [1, 2, 3]
                     my_list_2.insert(1, 5)  # Insert 5 at index 1
                     # Result: [1, 5, 2, 3]


    
    
    ```

---


11. **Q: What are the two methods for removing items from a list?**
   - **Ans:** The two methods for removing items from a list are `pop()` and `remove()`. 

   - The `pop()` method removes and returns the element at a specified index. If no index is provided, it removes and returns the last element by default.

   - The `remove()` method removes the first occurrence of a specified value from the list. It does not require an index but needs the value to be removed.

For Example:
```python
                              my_list = [1, 2, 3, 4, 5]

                              # Using pop() to remove the element at index 2
                              removed_item = my_list.pop(2)
                              # Now my_list is [1, 2, 4, 5] and removed_item is 3

                              # Using remove() to remove the value 4
                              my_list.remove(4)
                              # Now my_list is [1, 2, 5]
                              
```

12. **Describe how list values and string values are identical.**
    - **Ans:**  Both list values and string values are sequences of individual elements in Python. Elements in a list can be of any data type, and they are enclosed in square brackets [...], while elements in a string are characters and are enclosed in single quotes '...' or double quotes "...". Lists and strings are both ordered collections, which means the order of elements matters. Additionally, both lists and strings support indexing and slicing operations to access specific elements or subsequences.
